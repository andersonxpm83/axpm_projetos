```python
import pandas as pd
```

```python
# Carregar o arquivo CSV
file_path = "/Users/anderson.paiva/Downloads/results-20250414-155907 - results-20250414-155907.csv"
df = pd.read_csv(file_path)
```

```python
# Primeiro, converter a coluna 'pets_atendidos' e 'distance' (se for usada depois) para numérica
df['pets_atendidos'] = pd.to_numeric(df['pets_atendidos'], errors='coerce')

# Selecionar colunas de interesse
cols_interesse = ['pets_atendidos', 'tx_ocupacao', 'variacao_preco', 'csat_geral', 'pctual_glosa']

# Agrupar por 'regiao_atendimento' e 'classificacao_prestador' e calcular média, mediana e desvio padrão
agg_stats = df.groupby('classificacao_prestador')[cols_interesse]\
    .agg(['mean', 'median', 'std'])\
    .round(2)

agg_stats
```
